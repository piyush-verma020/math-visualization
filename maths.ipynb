```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, expon, uniform

# Set style
sns.set(style="whitegrid")

# Create a figure with subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# --------- 1. Normal Distribution ---------
x = np.linspace(-5, 5, 1000)
y = norm.pdf(x, loc=0, scale=1)  # Mean=0, Std=1

axs[0].plot(x, y, color='blue', lw=2)
axs[0].fill_between(x, y, alpha=0.3)
axs[0].set_title("Normal Distribution\n(mean=0, std=1)")
axs[0].set_xlabel("x")
axs[0].set_ylabel("Probability Density")

# --------- 2. Exponential Distribution ---------
x = np.linspace(0, 10, 1000)
y = expon.pdf(x, scale=1)  # lambda = 1  scale = 1

axs[1].plot(x, y, color='green', lw=2)
axs[1].fill_between(x, y, alpha=0.3)
axs[1].set_title("Exponential Distribution\n(lambda=1)")
axs[1].set_xlabel("x")
axs[1].set_ylabel("Probability Density")

# --------- 3. Uniform Distribution ---------
x = np.linspace(-1, 2, 1000)
y = uniform.pdf(x, loc=0, scale=1)  # from 0 to 1

axs[2].plot(x, y, color='orange', lw=2)
axs[2].fill_between(x, y, alpha=0.3)
axs[2].set_title("Uniform Distribution\n(from 0 to 1)")
axs[2].set_xlabel("x")
axs[2].set_ylabel("Probability Density")

plt.suptitle("Visualization of Common Probability Distributions", fontsize=16)
plt.tight_layout()
plt.show()
```
